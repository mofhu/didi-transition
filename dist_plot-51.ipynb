```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("data/season_1/groupby_all.csv")
```

```python
df.dtypes
```

```python
df[["start_district_num", "order_count", 'succeed_count', 'gap']] = df[["start_district_num", "order_count", 'succeed_count', 'gap']].\
    apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
df.dtypes
```

```python
dist_id = 51
```

```python
df_dist = df.loc[df['start_district_num'] == dist_id]
```

```python
df_dist
```

```python
df_dist.groupby('date').sum()
```

```python
df_dist.groupby('date').sum().plot()
```

```python
df_dist['time_slice'] = df['time_slice'].\
    apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
train_dates = []
for i in range(1, 22):
    if i < 10:
        train_dates.append('2016-01-0' + str(i))
    else:
        train_dates.append('2016-01-' + str(i))
print(train_dates)
df_dist.loc[df['date'].isin(train_dates)].groupby('time_slice').sum()
```

```python
df_dist.loc[df['date'].isin(train_dates)].groupby('time_slice').sum().plot()
```
