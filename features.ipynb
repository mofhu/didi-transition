```python
df1["date"] = df1["date"].apply(lambda x: pd.to_datetime(x, errors='coerce'))
```

```python
df1[["order_count", 'succeed_count', 'gap']] = \
df1[["order_count", 'succeed_count', 'gap']].astype(float)
# transfer object to numbers
```

```python
df_train_1 = df1[df.start_district_num == 1][df.date < '2016-01-22']
df_test_1 = df1[df.start_district_num == 1][df.date > '2016-01-21']
```

```python
df_train_1
```

```python
df_test_1
```

```python
X = df_train_1.as_matrix(columns=['time_slice','order_count','succeed_count',
                                  'weekday', 'is_holiday'])
y = df_train_1.as_matrix(columns=['gap']).ravel()
```

```python
X
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import cross_validation

X_train, X_test, y_train, y_test = cross_validation.train_test_split(
    X, y, test_size=0.2, random_state=0)

clf = RandomForestRegressor(n_estimators=5)
clf = clf.fit(X_train, y_train)

print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
from sklearn.cross_validation import cross_val_score

from sklearn.metrics.scorer import make_scorer
import math

def MAPE_scorer(y, y_pred):
    error = 0
    num = len(y)
    for i in range(0, num):
        if y[i] > 0:
            error += math.fabs(y_pred[i] - y[i]) / y[i]
    print(error, num)
    if num > 0:
        return error / num
    else:
        return 0

my_scorer = make_scorer(MAPE_scorer)

cross_val_score(clf, X=X_test, y=y_test, scoring=my_scorer, cv=None, n_jobs=1, verbose=0, fit_params=None, 
                pre_dispatch='2*n_jobs')
```

```python
#print(clf.predict(X_test))
#print(y_test)
print(MAPE_scorer(y_train, clf.predict(X_train)))
print(MAPE_scorer(y_test, clf.predict(X_test)))
```
