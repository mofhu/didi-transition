```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("data/season_1/groupby_all.csv")
```

```python
df.dtypes
```

```python
df[["start_district_num", "order_count", 'succeed_count', 'gap']] = df[["start_district_num", "order_count", 'succeed_count', 'gap']].\
    apply(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
df.dtypes
```

```python
dist_subtotal = df.groupby("start_district_num").sum()
```

```python
dist_subtotal["gap_ratio"] = dist_subtotal['gap'] / dist_subtotal['order_count']
```

```python
dist_subtotal.sort_values(by='order_count', ascending=False)
```

```python
dist_subtotal.sort_values(by='gap', ascending=False)
```

```python
dist_subtotal.sort_values(by='gap_ratio', ascending=False)
```

```python
dist_subtotal.sum()
```

```python
dist_subtotal.sort_index()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
plt.figure(); dist_subtotal.plot.bar(); plt.legend(loc='best'); 
```
